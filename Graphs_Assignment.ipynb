```python
import plotly
import plotly.express as px
import plotly.figure_factory as ff
import pandas as pd
import numpy as np
from plotly.express import data
import matplotlib.pyplot as plt
from matplotlib import style
from pathlib import Path
import logging
```

```python
pp=pd.read_csv('Superstore1.csv')
```

```python
pp1 = pp[(pp['Segment'] == 'Consumer')]
[['Order Date', 'Segment', 'State', 'City',
  'Category', 'Sales', 'Profit']]
pp1
```

```python
bd=pd.DataFrame(pp1.groupby(['Category','State']).agg({'Sales':'mean'})).reset_index()
bd
```

```python
#1
x = bd['Category']
y1 = bd['Sales']
plt.bar(x, y1, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Sales')
plt.title('Consumer Data')
plt.show()
```

```python
#2
plt.figure(figsize=(12, 6))
x = bd['State']
y = bd['Sales']
plt.scatter(x, y,label='Consumer Segment')
plt.xlabel('State')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.legend()
```

```python
#5
pp2=pp1.head(15)
df=pp2.groupby(['City']).agg({'Order Date':'count'}).reset_index()
plt.style.use('ggplot')
plt.figure(figsize=(12,6))
x=df['City']
y=df['Order Date']
plt.plot(x, y, 'black', linewidth=2, label='Total Orders')
v= np.argmax(y)
w = np.argmin(y)

plt.annotate('MAX', xy=(x[v],y[v]), xytext=(x[v], y[v]+0.5),
            arrowprops=dict(facecolor='blue'))
plt.annotate('MIN', xy=(x[w], y[w]), xytext=(x[w], y[w]+0.5),
            arrowprops=dict(facecolor='red'))


plt.xlabel('Cities')
plt.ylabel('No of Orders')
plt.xticks(rotation=90)
plt.legend()
```

```python
#6
pp2 = pp1.head(15)
df = pp2.groupby(['State']).agg({'Order Date': 'count'}).reset_index()

plt.figure(figsize=(12, 6))
x = df['State']
y = df['Order Date']

plt.bar(x, y, label='Total Orders', color='skyblue', edgecolor='black')
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.title('Order Count by State')
plt.xticks(rotation=90)
plt.legend()
plt.show()


```

```python
#7
subcategory_quantity = pp.groupby('Sub-Category')['Quantity'].sum().sort_values()
plt.figure(figsize=(8, 8))
subcategory_quantity.plot(kind='barh', color='lightcoral', edgecolor='black')
plt.title('Quantity Sold by Sub-Category', fontsize=14)
plt.xlabel('Total Quantity Sold', fontsize=12)
plt.ylabel('Sub-Category', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#8
plt.figure(figsize=(8, 5))
plt.scatter(pp['Discount'], pp['Profit'], alpha=0.6, color='purple')
plt.title('Discount vs. Profit', fontsize=14)
plt.xlabel('Discount', fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#9 Profit Margin by Category (Bar Chart)
pp['Profit Margin'] = pp['Profit'] / pp['Sales'] * 100
category_margin = pp.groupby('Category')['Profit Margin'].mean()
plt.figure(figsize=(8, 6))
category_margin.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Average Profit Margin by Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Profit Margin (%)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 3. Trend of Sales Over Time (Line Chart)
sales_over_time = pp.groupby('Order Date')['Sales'].sum().sort_index()
plt.figure(figsize=(12, 6))
plt.plot(sales_over_time.index, sales_over_time.values, color='darkorange', linewidth=2)
plt.title('Sales Trend Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```

Using Plotly

```python
pp['Order Date'] = pd.to_datetime(pp['Order Date'], errors='coerce')

# 1.Bar Chart
fig1 = px.bar(pp.groupby('Region')['Sales'].sum().reset_index(),
              x='Region', y='Sales', color='Region',
              title='Total Sales by Region')
fig1.show()
```

```python
# 2. Pie Chart
fig2 = px.pie(pp.groupby('Ship Mode')['Quantity'].sum().reset_index(),
              names='Ship Mode', values='Quantity',
              title='Quantity Distribution by Ship Mode')
fig2.show()
```

```python
# 3.Sunburst Chart
fig3 = px.sunburst(pp, path=['Category', 'Sub-Category'], values='Sales',
                   title='Sales by Category and Sub-Category',
                   color='Sales', color_continuous_scale='RdBu')
fig3.show()
```

```python
# 4. Scatter Plot
fig4 = px.scatter(pp, x='Sales', y='Profit', color='Category',
                  title='Sales vs. Profit by Category')
fig4.show()
```

```python
# 5.Bubble Chart
fig5 = px.scatter(pp, x='Discount', y='Profit', size='Sales', color='Sub-Category',
                  title='Discount vs. Profit (Bubble Chart)')
fig5.show()
```

```python
# 6. Line Chart
pp['Month-Year'] = pp['Order Date'].dt.to_period('M')
monthly_sales = pp.groupby('Month-Year')['Sales'].sum().reset_index()
monthly_sales['Month-Year'] = monthly_sales['Month-Year'].astype(str)
fig6 = px.line(monthly_sales, x='Month-Year', y='Sales',
               title='Monthly Sales Trend', markers=True)
fig6.show()
```

```python
# 7. Bar Chart
avg_discount = pp.groupby('Category')['Discount'].mean().reset_index()
fig7 = px.bar(avg_discount, x='Category', y='Discount', color='Category',
              title='Average Discount by Category')
fig7.show()
```

```python
# 8. Horizontal Bar Chart
subcategory_quantity = pp.groupby('Sub-Category')['Quantity'].sum().reset_index()
fig8 = px.bar(subcategory_quantity, y='Sub-Category', x='Quantity', orientation='h',
              title='Quantity Sold by Sub-Category', color='Sub-Category')
fig8.show()
```

```python
# 9.Bar Chart
top_customers = pp.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
fig9 = px.bar(top_customers, x='Customer Name', y='Sales',
              title='Top 10 Customers by Sales', color='Sales')
fig9.show()
```

```python
# 10. Stacked Bar Chart
region_segment_sales = pp.groupby(['Region', 'Segment'])['Sales'].sum().reset_index()
fig10 = px.bar(region_segment_sales, x='Region', y='Sales', color='Segment',
               title='Sales Contribution by Region and Segment', barmode='stack')
fig10.show()
```
